# Predicting Heart Disease
The WHO estimates that 17.9 million people die from cardiovascular diseases (CVDs) every year. There are multiple risk factors that contribute to CVD. The goal of this project is to accurately predict the likelihood of a new patient having heart disease in the future based on these risk factors.

## The Dataset
The dataset is anonymized patient data from multiple hospitals on several patients. The dataset contains the following features:

1.  Age: Age of patient [years]
2.  Sex: Sex of patient [M :Male, F: Female]
3.  ChestPainType: Chest pain type [TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic]
4.  RestingBP: Resting blood pressure [mm Hg]
5.  Cholesterol: Serum Cholesterol [mm/dl]
6.  FastingBS: Fasting blood sugar [1: if FastingBS > 120 mg/dl, 0: otherwise]
7.  RestingECG: Resting ECG results [Normal: Normal, ST: ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), LVH: Probable or definite left ventricular hypertrophy by Estes' criteria]
8.  MaxHR: Max heart rate [60-202]
9.  ExerciseAngina: Exercise-induced angina [Y: Yes, N: No]
10. OldPeak: Oldpeak = ST [Numeric value measured in depression]
11. ST_Slope: Slope of the peak exercise ST segment [Up: upsloping, Flat: flat, Down: downsloping]
12. HeartDisease: Output Class [1: heart disease, 0: Normal]


```python
# Library Import
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
# Reading in the dataset
hd = pd.read_csv('heart_disease_prediction.csv')
print(hd.head())
```

```python
# Exploring the dataset
print('Number of observations:', hd.shape[0])
print('Number of features:', hd.columns.size - 1)
```

```python
# Descriptive Statistics of the dataset
print(hd.dtypes)
print(hd.info())
```

## Data Cleaning
Based on the observations from above, the columns RestingBP and Cholesterol need attention. Neither of these values can have a realistic minimum of 0. If there are relatively few observations with these errors, we can remove these observations from the dataset with minimal impact to the performance of the model.

```python
# Determining how many observations have RestingBP = 0 or Cholesterol = 0
print('Number of observations with 0 RestingBP:', hd['RestingBP'].value_counts()[0])
print('Number of observations with 0 Cholesterol:', hd['Cholesterol'].value_counts()[0])
```

The number of observations with RestingBP = 0 is only one, therefore it can be safely removed. However, the number of observations with Cholesterol = 0 is 172. This is too large of a number to justify removing from the dataset. Therefore, we will replace these values with the mean of Cholesterol based on whether the patient has heart disease or not.

```python
# Dropping Observation where RestingBP = 0
hd[hd['RestingBP'] == 0].index
hd.drop(hd[hd['RestingBP'] == 0].index, inplace=True)

# Creating copy of dataset to drop Cholesterol=0 rows in order to assess accurate mean
hd_copy = hd.copy()
hd_copy.drop(hd_copy[hd_copy['Cholesterol'] == 0].index, inplace=True)

# Determining average cholesterol for heart disease and for no heart disease
avg_chol_hd = hd_copy.groupby('HeartDisease').Cholesterol.mean()[1]
avg_chol_no_hd = hd_copy.groupby('HeartDisease').Cholesterol.mean()[0]
print('Average Cholesterol With Heart Disease:', avg_chol_hd)
print('Average Cholesterol With No Heart Disease:', avg_chol_no_hd)

# Replacing zero values with respective means based on HeartDisease or not
chol_0_hd_1_index = hd[(hd['Cholesterol'] == 0) & (hd['HeartDisease'] == 1)].index
chol_0_hd_0_index = hd[(hd['Cholesterol'] == 0) & (hd['HeartDisease'] == 0)].index
hd.loc[chol_0_hd_1_index, 'Cholesterol'] = avg_chol_hd
hd.loc[chol_0_hd_0_index, 'Cholesterol'] = avg_chol_no_hd
```

## Feature Selection
Now we will convert the categorical features into dummy variables so that we are able to train the model on them. We will also calculate the Pearson's correlation to determine which features are reasonably correlated to HeartDisease.



```python
hd_copy = hd.copy()
hd_copy = pd.get_dummies(data=hd_copy, drop_first=True)

# Determining correlations
correlations = abs(hd_copy.corr())
plt.figure(figsize=(12,8))
sns.heatmap(correlations, annot=True, cmap='Blues')
```

```python
# Highest Correlated Features with Heart Disease
plt.figure(figsize=(12,8))
sns.heatmap(correlations[correlations > 0.4], annot=True, cmap='Blues')
plt.show()

# Selected Features based off a threshold of 0.4
correlated_features = ['MaxHR', 'Oldpeak', 'ChestPainType_ATA', 'ExerciseAngina_Y', 'ST_Slope_Flat', 'ST_Slope_Up']
```

## Building The Classifier
The classifier of interest for this project is k-NN. The data will be split into training and validation sets. The classifier will be built with one feature at a time to determine the efficacy of the feature.

```python
# Splitting features from target
X = hd_copy.drop('HeartDisease', axis=1)
y = hd_copy['HeartDisease']

# Using only the selected features determined from the correlation above (Model performs better with all features included)
#X = X[correlated_features]

# Splitting the dataset into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state = 77)

# Normalizing Dataset
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
# Constructing the classifier using the correlated features and using Grid Search to optimize hyperparameters
grid_params = {"n_neighbors": range(1, 200),
               'weights': ['uniform', 'distance'],
                "metric": ["minkowski", "manhattan"]
              }
knn = KNeighborsClassifier()
knn_grid = GridSearchCV(knn, grid_params, scoring='accuracy')
knn_grid.fit(X_train_scaled, y_train)

# Viewing best score and parameters achieved
print('Best Score:', knn_grid.best_score_)
print('Best Parameters:', knn_grid.best_params_)

# Obtaining best classifier and applying to validation data
best_estimator = knn_grid.best_estimator_
accuracy = best_estimator.score(X_val_scaled, y_val)

# Viewing Accuracy achieved with model
print('Accuracy:', accuracy)
```

## Conclusions
We can see that our k-NN classifier performs well, with an accuracy of 85.87%. This accuracy is obtained when using all feature columns available to us instead of selecting the few that have high correlations to heart disease. This adds computational complexity, but given the seriousness of diagnosing someone with heart disease, accuracy should be preferred over speed.
